```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Flatten,Dense
```

```python
(X_train,Y_train),(x_test,y_test) = fashion_mnist.load_data()
```

```python
display(X_train.shape)
display(x_test.shape)
```

```python
Labels= ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat","Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
plt.figure(figsize=(4,4))
for i in range (25):
    plt.subplot(5,5,i+1)
    plt.imshow(X_train[i])
    plt.title(f"{Labels[Y_train[i]]}",fontsize=8)
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
X_train,x_test = X_train/255,x_test/255
```

```python
Y_train = to_categorical(Y_train,num_classes=10)
```

```python
X_train = X_train.reshape(X_train.shape[0],28,28,1)
x_test = x_test.reshape(x_test.shape[0],28,28,1)
```

```python
cnn = Sequential([Conv2D(32,(2,2),activation='relu',input_shape=(28,28,1)),
                  MaxPooling2D((2,2)),
                  Conv2D(64,(2,2),activation='relu'),
                  MaxPooling2D((2,2)),
                  Flatten(),
                  Dense(64,activation='relu'),
                  Dense(64,activation='relu'),
                  Dense(10,activation='softmax')])
```

```python
cnn.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
cnn.fit(X_train,Y_train,epochs=10)
```

```python
a= X_train[0].reshape(1,28,28,1)
Labels[np.argmax(cnn.predict(a))]
```

```python
plt.figure(figsize=(3,3))
plt.imshow(X_train[0])
plt.title(f'{Labels[np.argmax(cnn.predict(a))]}')
plt.axis('off');
```
